```python
import numpy as np # type: ignore
import json
import matplotlib.pyplot as plt # type: ignore
```

```python
np.random.seed(42)
total=1000
```

```python
x=np.random.uniform(-400,400,total)
y=np.random.uniform(-400,400,total)
z=np.random.uniform(-400,400,total)
print(x[:3])
```

```python
points=[]
for n in range(total):
    points.append({"x":float(x[n]),"y":float(y[n]),"z":float(z[n])})

print(points[:3])
```

```python
with open("data.json","w")as file:
    data=json.dump(points,file,indent=4)
print(f"{total} data genrated sucessfully")
```
